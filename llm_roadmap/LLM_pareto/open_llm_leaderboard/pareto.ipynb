```python
def plot_pareto_frontier(Xs, Ys, maxX=True, maxY=True):
    '''Pareto frontier selection process'''
    sorted_list = sorted([[Xs[i], Ys[i]] for i in range(len(Xs))], reverse=maxY)
    pareto_front = [sorted_list[0]]
    for pair in sorted_list[1:]:
        if maxY:
            if pair[1] >= pareto_front[-1][1]:
                pareto_front.append(pair)
        else:
            if pair[1] <= pareto_front[-1][1]:
                pareto_front.append(pair)
    pareto_front.append([min(Xs),max(Ys)])
    
    '''Plotting process'''
    plt.scatter(Xs,Ys)
    pf_X = [pair[0] for pair in pareto_front]
    pf_Y = [pair[1] for pair in pareto_front]
    plt.plot(pf_X, pf_Y)
    plt.xlabel("Objective 1")
    plt.ylabel("Objective 2")
    plt.show()
    return pareto_front

# Example usage:
Xs = [3, 1, 4, 2, 8]
Ys = [8, 4, 2, 6, 1]

front = plot_pareto_frontier(Xs, Ys, maxX=True, maxY=True)
front

```

```python
def plot_pareto_frontier(Xs, Ys, connect_to_origin=True):
    # Combine X and Y values into a list of tuples
    data = list(zip(Xs, Ys))
    
    # Pareto frontier selection process
    pareto_front = []

    for i, (x, y) in enumerate(data):
        # Check if (x, y) is dominated by any point in the Pareto frontier
        if not any(x2 >= x and y2 >= y for x2, y2 in pareto_front):
            pareto_front.append((x, y))
    
    # Sort the Pareto frontier by x-values
    pareto_front = sorted(pareto_front, key=lambda point: point[0])

    # Connect to the origin if specified
    if connect_to_origin:
        pareto_front.insert(0, (min(Xs), pareto_front[0][1]))
    
    # Plotting process
    plt.scatter(Xs, Ys)
    pf_X = [pair[0] for pair in pareto_front]
    pf_Y = [pair[1] for pair in pareto_front]
    plt.plot(pf_X, pf_Y)
    plt.xlabel("Objective 1")
    plt.ylabel("Objective 2")
    plt.show()
    return pareto_front

```

```python
import matplotlib.pyplot as plt

def find_pareto_frontier(x_values, y_values, xmax=True, ymax=True):
    """
    Find Pareto frontier given x and y values along with criteria for maximization.

    Parameters:
    - x_values: List of x values
    - y_values: List of y values
    - xmax: Boolean indicating whether to maximize x (True) or minimize x (False)
    - ymax: Boolean indicating whether to maximize y (True) or minimize y (False)

    Returns:
    - List of points on the Pareto frontier
    """

    pareto_frontier = []
    # Combine x and y values into a list of tuples
    points = list(zip(x_values, y_values))

    x_minimum = min(x_values)
    x_maximum = max(x_values)
    y_minimum = min(y_values)
    y_maximum = max(y_values)

    # Sort points based on x and y criteria
    # Add first point
    if xmax == True and ymax == True:
        points.sort(key=lambda point: (-point[0], -point[1]))  # Sort by x descending, y descending
        pareto_frontier.append([x_maximum, y_minimum])
    elif xmax == True and ymax == False:
        points.sort(key=lambda point: (-point[0], point[1]))  # Sort by x descending, y ascending
        pareto_frontier.append([x_maximum, y_maximum])
    elif xmax == False and ymax == True:
        points.sort(key=lambda point: (point[0], -point[1]))  # Sort by x ascending, y descending
        pareto_frontier.append([x_minimum, y_minimum])
    else:
        points.sort(key=lambda point: (point[0], point[1]))  # Sort by x ascending, y ascending
        pareto_frontier.append([x_minimum, y_maximum])

    # now loop to find the optimimum
    while len(points) != 0:
        # Initialize the Pareto frontier with the current max point
        current_x = points[0][0]
        current_y = points[0][1]
        pareto_frontier.append([current_x,current_y])

        # Iterate through the sorted points to remove any points lower in x
        local_points = []
        for point in points:
            if (xmax and point[0] > current_x) or (not xmax and point[0] < current_x):
                local_points.append(point)
            elif (ymax and point[1] > current_y) or (not ymax and point[1] < current_y):
                local_points.append(point)
        points = local_points

    # Add last point
    if xmax == True and ymax == True:
        pareto_frontier.append([x_minimum, y_maximum])
    elif xmax == True and ymax == False:
        pareto_frontier.append([x_minimum, y_minimum])
    elif xmax == False and ymax == True:
        pareto_frontier.append([x_maximum, y_maximum])
    else:
        pareto_frontier.append([x_maximum, y_minimum])


    # Separate x and y values of the Pareto frontier for plotting
    pareto_x, pareto_y = zip(*pareto_frontier)

    # Plot the input points and the Pareto frontier
    # square plot
    #plt.axis('square')
    fig1, ax = plt.subplots()
    ax.set_box_aspect(1)
    ax.scatter(x_values, y_values, label='Input Points')
    ax.plot(pareto_x, pareto_y, label='Pareto Frontier', color='red')
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_title('Pareto Frontier Plot')
    ax.legend()
    plt.show()

    return pareto_frontier

# Example usage:
x_values = [3, 1, 4, 2, 6, 8]
y_values = [2, 5, 1, 6, 4, 3]

pareto_frontier = find_pareto_frontier(x_values, y_values, xmax=True, ymax=True)
pareto_frontier = find_pareto_frontier(x_values, y_values, xmax=True, ymax=False)
pareto_frontier = find_pareto_frontier(x_values, y_values, xmax=False, ymax=True)
pareto_frontier = find_pareto_frontier(x_values, y_values, xmax=False, ymax=False)


```
